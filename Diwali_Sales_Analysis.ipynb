## EDA

Gender

```python
plt.figure(figsize=[10,4])
ax = sns.countplot(x='Gender',data= df ,hue ='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=[10,3])
Gender_Sales = df.groupby(['Gender'],as_index =False)['Amount'].sum().sort_values(by='Amount',ascending = False)

sns.barplot(x='Gender',y= 'Amount',data=Gender_Sales,hue = 'Gender')


plt.show()
```

From the above graphs , We can see that Most of the buyers are Females and even the purchasing power of Females is greater than Males 

## Age

```python
plt.figure(figsize=[10,4])
ax = sns.countplot(x = 'Age Group',data = df,hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
plt.figure(figsize=[15,4])
ax = sns.barplot(x= 'Age Group',y ='Amount',hue = 'Gender',data = df ,estimator = 'mean')

for bars in ax.containers:
    ax.bar_label(bars,fmt ='%.1f',label_type='edge')
```

```python
plt.figure(figsize=[10,4])

Sales_Age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x ='Age Group', y='Amount',data = Sales_Age,hue = 'Age Group' )
```

From above graphs , we observe that most of the buyers are of age group between 26-35 yrs female

## State-wise

```python
# sns.set(rc={'figure.figsize':(25,10)})
plt.figure(figsize=[25,10])
State_Amount = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(data = State_Amount,x='State',y = 'Amount',hue = 'State')

plt.show()
```

From the above Graphs , we can see that most of the Orders & total sales/ Amount are from Uttar Pradesh , Maharasthtra and Karnataka respectively....

## Occupation

```python
# sns.set(rc={'figure.figsize':(18,5)})
plt.figure(figsize=[18,5])
ax = sns.countplot(x='Occupation',data = df,hue = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# sns.set(rc={'figure.figsize':(20,5)})
plt.figure(figsize=[20,5])

Occupation_Amount = df.groupby(['Occupation'],as_index= False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(15)

sns.barplot(data= Occupation_Amount,x = 'Occupation',y = 'Amount',hue = 'Occupation')

plt.show()
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

## Product Category

```python
df.columns
```

```python
# sns.set(rc={'figure.figsize':(30,10)})
plt.figure(figsize=[30,10])

ax = sns.countplot(x='Product_Category',data = df)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
Product_Sales = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# sns.set(rc={'figure.figsize':(20,5)})
plt.figure(figsize=[20,5])
sns.barplot(data = Product_Sales, x = 'Product_Category',y= 'Amount',hue = 'Product_Category')
plt.show()
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
Id_Sales = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# sns.set(rc={'figure.figsize':(20,5)})
plt.figure(figsize=[20,5])
sns.barplot(data = Id_Sales, x = 'Product_ID',y= 'Orders',hue ='Product_ID')
plt.show()
```

## Conclusion

Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
